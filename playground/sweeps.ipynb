```python
import os


from omegaconf import OmegaConf
if os.path.basename(os.getcwd()) == "playground":
    os.chdir("..")

print(f"cwd: {os.getcwd()}")
from sweep import RandomSweep


from cores.core.values.constants import Datasets, GNNLayers

DATASET_LIST = []
for dataset_name, dataset_value in vars(Datasets).items():
    if not dataset_name.startswith("__"):
        DATASET_LIST.append(dataset_value)

```

```python
cfg = OmegaConf.create()

cfg.machine_file_path = os.path.join("config", "machine", "mps.yaml")
cfg.logger_file_path = os.path.join("config", "logger", "file_system.yaml")
cfg.dataset_file_path = os.path.join("config", "dataset", "mutag.yaml")
```

# GNN Sweeps

```python
gnn_sweep_file_path=os.path.join("config", "sweep", "gnn.yaml")




for dataset in DATASET_LIST:
    
    for archi in list(GNNLayers):
        param_files = []
        archi_file_path = os.path.join("config", "sweep", f"archi_{archi.value}.yaml")
        param_files.append(archi_file_path)
        dataset_file_path = os.path.join("config", "sweep", f"dataset_{dataset}.yaml")
        param_files.append(dataset_file_path)
        sweep = RandomSweep(sweep_file_path=gnn_sweep_file_path, param_files = param_files)

        configs = sweep.sample(seed=0, count=5)

        command_main = "python cores/delivery/cli/cores_train.py "

        for i, config in enumerate(configs):
            print(f"config {i}:\n\t{config}")
```
